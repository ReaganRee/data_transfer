```python
def seq2kmer(seq, k):
    """
    Convert original sequence to kmers

    Arguments:
    seq -- str, original sequence.
    k -- int, kmer of length k specified.

    Returns:
    kmers -- str, kmers separated by space

    """
    kmer = [seq[x:x+k] for x in range(len(seq)+1-k)]
    kmers = " ".join(kmer)
    return kmers



from Bio.SeqRecord import SeqRecord
from Bio import SeqIO
from tqdm import tqdm
import csv

def process_fa_to_tsv_train(rbp):
  for i in range(5):
    kmer_to_tsv = []
    input = "/content/drive/MyDrive/all_data_process/22_RBP_data/22_RBP/{rbp1}/{fold1}/fold_{fold2}_train.fa".format(rbp1=rbp,fold1=i,fold2=i)

    for seq_record in tqdm(SeqIO.parse(input, "fasta")):
      seq_record.seq = seq_record.seq.upper()
      final_kmer = seq2kmer(str(seq_record.seq), 1)
      label = seq_record.id
      kmer_to_tsv.append([final_kmer, label])
      #print([final_kmer + '\t' + label])


    with open('/content/drive/MyDrive/all_data_process/22_RBP_data/rna_fm/{rbp1}/{fold}/train_val/train.tsv'.format(rbp1=rbp,fold=i), 'w') as f:
        tsv_w = csv.writer(f, delimiter='\t')
        tsv_w.writerow(['sequence', 'label'])
        for row in kmer_to_tsv:
          tsv_w.writerow(row)





def process_fa_to_tsv_vali(rbp):
  for i in range(5):
    kmer_to_tsv = []
    input = "/content/drive/MyDrive/all_data_process/22_RBP_data/22_RBP/{rbp1}/{fold1}/fold_{fold2}_val.fa".format(rbp1=rbp,fold1=i,fold2=i)

    for seq_record in tqdm(SeqIO.parse(input, "fasta")):
      seq_record.seq = seq_record.seq.upper()
      final_kmer = seq2kmer(str(seq_record.seq), 1)
      label = seq_record.id
      kmer_to_tsv.append([final_kmer, label])
      #print([final_kmer + '\t' + label])


    with open('/content/drive/MyDrive/all_data_process/22_RBP_data/rna_fm/{rbp1}/{fold}/train_val/dev.tsv'.format(rbp1=rbp,fold=i), 'w') as f:
        tsv_w = csv.writer(f, delimiter='\t')
        tsv_w.writerow(['sequence', 'label'])
        for row in kmer_to_tsv:
          tsv_w.writerow(row)


def process_fa_to_tsv_test(rbp):
  for i in range(5):
    kmer_to_tsv = []
    input = "/content/drive/MyDrive/all_data_process/22_RBP_data/22_RBP/{rbp1}/test.fa".format(rbp1=rbp)

    for seq_record in tqdm(SeqIO.parse(input, "fasta")):
      seq_record.seq = seq_record.seq.upper()
      final_kmer = seq2kmer(str(seq_record.seq), 1)
      label = seq_record.id
      kmer_to_tsv.append([final_kmer, label])
      #print([final_kmer + '\t' + label])


    with open('/content/drive/MyDrive/all_data_process/22_RBP_data/rna_fm/{rbp1}/{fold}/test/dev.tsv'.format(rbp1=rbp,fold=i), 'w') as f:
        tsv_w = csv.writer(f, delimiter='\t')
        tsv_w.writerow(['sequence', 'label'])
        for row in kmer_to_tsv:
          tsv_w.writerow(row)

with open("/content/drive/MyDrive/all_data_process/22_RBP_data/22_RBP/22_all_list.txt", "r") as f:
      rbp_name_list = f.readlines()

for rbp in rbp_name_list:
  bare_name = rbp.strip()
  process_fa_to_tsv_train(bare_name)
  process_fa_to_tsv_vali(bare_name)
  process_fa_to_tsv_test(bare_name)




```

```python
from Bio import SeqIO
from tqdm import tqdm
def cut_fasta_sequence(sequence, chunk_size=510):
    chunks = []
    for i in range(0, len(sequence), chunk_size):
        chunk = sequence[i:i + chunk_size]
        chunks.append(chunk)
    return chunks

# Example usage:



# Print the chunks

seq_to_write = []

for record in tqdm(SeqIO.parse("/content/drive/MyDrive/审稿补实验/sub_cellular_localization_motif/5_fold_data/all_combined_test/all_combined_test_removed.fa", "fasta")):
  chunks = cut_fasta_sequence(record)
  #print(chunks)
  seq_to_write.extend(chunks)

count = 0
for seq in seq_to_write:
  seq.id = str(count)
  count += 1

SeqIO.write(seq_to_write, "/content/drive/MyDrive/审稿补实验/sub_cellular_localization_motif/5_fold_data/all_combined_test/all_combined_test_removed_chunked.fa", "fasta")
```

```python
def seq2kmer(seq, k):
    """
    Convert original sequence to kmers

    Arguments:
    seq -- str, original sequence.
    k -- int, kmer of length k specified.

    Returns:
    kmers -- str, kmers separated by space

    """
    kmer = [seq[x:x+k] for x in range(len(seq)+1-k)]
    kmers = " ".join(kmer)
    return kmers



from Bio.SeqRecord import SeqRecord
from Bio import SeqIO
from tqdm import tqdm
import csv




def process_fa_to_tsv_test():
  #for i in range(5):
    kmer_to_tsv = []
    input = "/content/drive/MyDrive/审稿补实验/sub_cellular_localization_motif/5_fold_data/all_combined_test/all_combined_test_removed_chunked.fa"



    for seq_record in tqdm(SeqIO.parse(input, "fasta")):
      seq_record.seq = seq_record.seq.upper()
      final_kmer = seq2kmer(str(seq_record.seq), 3)
      label = int(seq_record.id) % 2
      final_kmer = final_kmer.replace("T", "U")
      kmer_to_tsv.append([final_kmer, label])
      #print([final_kmer + '\t' + label])


    with open('/content/drive/MyDrive/审稿补实验/sub_cellular_localization_motif/5_fold_data/all_combined_test/dev.tsv', 'w') as f:
        tsv_w = csv.writer(f, delimiter='\t')
        tsv_w.writerow(['sequence', 'label'])
        for row in kmer_to_tsv:
          tsv_w.writerow(row)


process_fa_to_tsv_test()




```
